```python
import matplotlib.pyplot as plt

categories = ["food", "recharge", "stationery", "transport", "savings"]
amounts = [350, 199, 120, 180, 300]

colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"]

plt.figure(figsize=(7, 7))
plt.pie(
    amounts,
    labels=categories,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("my pocket money breakdown")
plt.tight_layout()
plt.show()
```
