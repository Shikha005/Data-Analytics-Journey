```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "name": ["priya", "rahul", "sneha", "amit", "pooja"],
    "marks": [85, 72, 90, 65, 78],
    "subject": ["maths", "english", "science", "hindi", "computer"]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.barplot(x="name", y="marks", data=df, palette="Blues_d")

plt.title("student marks")
plt.xlabel("student")
plt.ylabel("marks")
plt.tight_layout()
plt.show()
```
