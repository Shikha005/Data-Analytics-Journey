```python
import matplotlib.pyplot as plt

months = ["jan", "feb", "mar", "apr", "may"]
saved =  [800, 1200, 600, 1500, 900]
goal =   [1000, 1000, 1000, 1000, 1000]

plt.figure(figsize=(8, 5))

bars = plt.bar(months, saved, color=[
    "green" if s >= g else "coral"
    for s, g in zip(saved, goal)
])

plt.plot(months, goal, color="blue",
         linestyle="--", label="goal rs.1000")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 20,
        str(int(bar.get_height())),
        ha="center"
    )

plt.title("monthly savings tracker")
plt.ylabel("amount saved (rs)")
plt.legend()
plt.tight_layout()
plt.show()
```
