```python
students = {
    "priya": 88,
    "rahul": 72,
    "sneha": 95,
    "amit": 61,
    "pooja": 79
}

def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "F"

with open("results.txt", "w") as f:
    f.write("student result card\n")
    f.write("-" * 30 + "\n")
    for name in students:
        grade = get_grade(students[name])
        line = name + " | " + str(students[name]) + " | " + grade
        f.write(line + "\n")
    f.write("-" * 30 + "\n")

print("results saved to file!")

print("\nreading saved file:")
print("-" * 30)

with open("results.txt", "r") as f:
    content = f.read()
    print(content)
```
