```python
items = ["Rice", "Dal", "Oil", "Sugar", "Tea"]
prices = [120, 90, 180, 50, 75]
quantities = [2, 1, 1, 3, 2]

total = 0

print("my shopping bill")
print("-" * 30)

for i in range(len(items)):
    cost = prices[i] * quantities[i]
    total = total + cost
    print(items[i], "->", cost)

print("-" * 30)
print("total =", total)

if total > 500:
    discount = total * 0.10
    print("got 10% discount -", discount)
    print("final total =", total - discount)
else:
    print("no discount today")
```
