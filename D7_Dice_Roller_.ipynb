```python
import random

print("dice roller")
print("-" * 20)

for i in range(5):
    dice = random.randint(1, 6)
    print("roll", i+1, "->", dice)

print("-" * 20)

total = 0
for i in range(10):
    total = total + random.randint(1, 6)

print("total of 10 rolls:", total)
print("average:", total / 10)
```
