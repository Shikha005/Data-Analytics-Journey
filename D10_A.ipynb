```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "product": ["chips", "cold drink", "biscuits", "chocolate", "juice", "water"],
    "category": ["snacks", "drinks", "snacks", "snacks", "drinks", "drinks"],
    "price": [20, 40, 30, 50, 60, 20],
    "units_sold": [150, 200, 120, 80, 90, 250],
    "month": ["jan", "jan", "jan", "jan", "jan", "jan"]
}

df = pd.DataFrame(data)

df["revenue"] = df["price"] * df["units_sold"]

print("shop sales dashboard")
print("-" * 50)
print(df[["product", "units_sold", "revenue"]])
print("-" * 50)
print("total revenue: Rs.", df["revenue"].sum())
print("best product:", df[df["revenue"] == df["revenue"].max()]["product"].values[0])
print("-" * 50)

cat_rev = df.groupby("category")["revenue"].sum()
print("category wise revenue:")
print(cat_rev)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].barh(df["product"], df["revenue"], color="steelblue")
axes[0].set_title("revenue by product")
axes[0].set_xlabel("revenue (rs)")

axes[1].pie(
    df["units_sold"],
    labels=df["product"],
    autopct="%1.0f%%",
    startangle=90
)
axes[1].set_title("units sold share")

axes[2].bar(cat_rev.index, cat_rev.values, color=["coral", "green"])
axes[2].set_title("revenue by category")
axes[2].set_ylabel("revenue (rs)")

plt.suptitle("january shop dashboard", fontsize=14)
plt.tight_layout()
plt.show()
```
