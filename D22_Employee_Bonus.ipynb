```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def bonus(self):
        return 0

    def show(self):
        print(self.name, "| salary:", self.salary,
              "| bonus:", self.bonus())

class Developer(Employee):
    def bonus(self):
        return self.salary * 0.20

class Designer(Employee):
    def bonus(self):
        return self.salary * 0.15

class Manager(Employee):
    def bonus(self):
        return self.salary * 0.30

class Intern(Employee):
    def bonus(self):
        return 500

employees = [
    Developer("priya", 60000),
    Designer("rahul", 50000),
    Manager("sneha", 90000),
    Intern("amit", 15000)
]

for emp in employees:
    emp.show()
```
