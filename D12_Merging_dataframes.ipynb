```python
import pandas as pd

students = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105],
    "name": ["priya", "rahul", "sneha", "amit", "pooja"],
    "city": ["indore", "bhopal", "delhi", "mumbai", "indore"]
})

marks = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 106],
    "maths": [78, 65, 90, 55, 82],
    "english": [85, 72, 88, 60, 79],
    "science": [92, 68, 95, 58, 85]
})

print("students table:")
print(students)
print("-" * 35)
print("marks table:")
print(marks)
print("-" * 35)

inner = pd.merge(students, marks, on="student_id", how="inner")
print("inner join (only matching):")
print(inner)
print("-" * 35)

left = pd.merge(students, marks, on="student_id", how="left")
print("left join (all students):")
print(left)
print("-" * 35)

inner["total"] = inner["maths"] + inner["english"] + inner["science"]
inner["percentage"] = round((inner["total"] / 300) * 100, 2)

print("final merged data with totals:")
print(inner[["name", "city", "total", "percentage"]])
```
