```python
class Phone:
    def __init__(self, brand):
        self.brand = brand

    def camera(self):
        print(self.brand, "basic camera")

    def battery(self):
        print(self.brand, "normal battery")

class Samsung(Phone):
    def camera(self):
        print(self.brand, "108mp pro camera")  # overriding

class Iphone(Phone):
    def camera(self):
        print(self.brand, "48mp cinematic camera")  # overriding

    def battery(self):
        print(self.brand, "optimized battery life")  # overriding

class Redmi(Phone):
    def battery(self):
        print(self.brand, "6000mah massive battery")  # overriding

phones = [Samsung("samsung"), Iphone("iphone"), Redmi("redmi")]

for phone in phones:
    phone.camera()
    phone.battery()
    print("-" * 30)
```
