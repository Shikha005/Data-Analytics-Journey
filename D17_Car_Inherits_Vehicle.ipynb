```python
class Vehicle:

    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print(self.brand, "is moving at", self.speed, "kmph")

    def stop(self):
        print(self.brand, "stopped")

class Car(Vehicle):

    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)
        self.doors = doors

    def honk(self):
        print(self.brand, "says: beep beep!")

    def info(self):
        print("brand:", self.brand)
        print("speed:", self.speed, "kmph")
        print("doors:", self.doors)

c1 = Car("toyota", 120, 4)
c2 = Car("honda", 140, 2)

c1.info()
c1.move()
c1.honk()
c1.stop()
print("-" * 25)
c2.info()
c2.honk()
```
