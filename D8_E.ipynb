```python
import pandas as pd

data = {
    "name": ["priya", "rahul", "sneha", "amit", "pooja", "vikram", "neha", "raj"],
    "city": ["indore", "bhopal", "indore", "mumbai", "bhopal", "indore", "mumbai", "bhopal"],
    "marks": [85, 72, 90, 65, 78, 55, 88, 70],
    "fees": [5000, 4500, 5000, 6000, 4500, 5000, 6000, 4500]
}

df = pd.DataFrame(data)

print("student data")
print("-" * 40)
print(df)
print("-" * 40)

print("average marks by city:")
city_avg = df.groupby("city")["marks"].mean()
print(city_avg)
print("-" * 40)

print("students per city:")
city_count = df.groupby("city")["name"].count()
print(city_count)
print("-" * 40)

print("city wise summary:")
summary = df.groupby("city")["marks"].agg(["mean", "max", "min", "count"])
print(summary)
```
