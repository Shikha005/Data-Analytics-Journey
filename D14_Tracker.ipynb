```python
import matplotlib.pyplot as plt

days = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
mood = [6, 5, 7, 4, 8, 9, 7]

colors = []
for m in mood:
    if m >= 8:
        colors.append("green")
    elif m >= 6:
        colors.append("orange")
    else:
        colors.append("red")

plt.figure(figsize=(8, 5))
plt.bar(days, mood, color=colors)
plt.plot(days, mood, marker="o", color="black", linewidth=1)

plt.title("my mood this week")
plt.ylabel("mood score out of 10")
plt.ylim(0, 11)

for i in range(len(days)):
    plt.text(i, mood[i] + 0.2, str(mood[i]), ha="center")

plt.tight_layout()
plt.show()
```
