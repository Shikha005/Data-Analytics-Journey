```python
names = ["  priya sharma  ", "RAHUL verma", "sneha GUPTA",
         "amit KUMAR  ", "  pooja singh"]

print("cleaning messy names")
print("-" * 30)

for name in names:

    cleaned = name.strip()

    proper = cleaned.title()
    print("before:", repr(name))
    print("after: ", proper)
    print()
```
