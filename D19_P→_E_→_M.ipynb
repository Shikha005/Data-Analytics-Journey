```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print("name:", self.name, "| age:", self.age)

class Employee(Person):
    def __init__(self, name, age, company, salary):
        super().__init__(name, age)
        self.company = company
        self.salary = salary

    def work(self):
        print(self.name, "is working at", self.company)

    def show_salary(self):
        print("salary: Rs.", self.salary)

class Manager(Employee):
    def __init__(self, name, age, company, salary, team_size):
        super().__init__(name, age, company, salary)
        self.team_size = team_size

    def manage(self):
        print(self.name, "is managing", self.team_size, "people")

    def appraisal(self, percent):
        hike = self.salary * (percent / 100)
        self.salary = self.salary + hike
        print(percent, "% hike! new salary: Rs.", self.salary)

m = Manager("priya", 30, "tcs", 80000, 10)
m.introduce()
m.work()
m.show_salary()
m.manage()
m.appraisal(15)
```
