```python
contacts = {
    "priya": {"phone": "9876543210", "city": "indore"},
    "rahul": {"phone": "9812345678", "city": "bhopal"},
    "sneha": {"phone": "9856781234", "city": "delhi"},
    "amit": {"phone": "9890123456", "city": "mumbai"}
}

def show_contact(name):
    if name in contacts:
        print("name:", name)
        print("phone:", contacts[name]["phone"])
        print("city:", contacts[name]["city"])
    else:
        print("contact not found")

print("my contact book")
print("-" * 25)

for person in contacts:
    show_contact(person)
    print("-" * 25)
```
