```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "month": ["jan", "feb", "mar", "apr", "may", "jun",
               "jul", "aug", "sep", "oct", "nov", "dec"],
    "temp": [22, 26, 33, 40, 44, 41, 35, 33, 32, 30, 25, 21]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
sns.lineplot(x="month", y="temp", data=df,
             marker="o", color="red", linewidth=2)

plt.title("indore temperature 2026")
plt.ylabel("temperature (celsius)")
plt.axhline(y=35, color="orange", linestyle="--", label="heat warning")
plt.legend()
plt.tight_layout()
plt.show()
```
