```python
import matplotlib.pyplot as plt

week = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
healthy = [2, 1, 3, 2, 1, 0, 2]
junk =    [1, 2, 0, 1, 2, 3, 1]

plt.figure(figsize=(9, 5))
plt.plot(week, healthy, marker="o", color="green", label="healthy")
plt.plot(week, junk, marker="o", color="red", label="junk food")

plt.title("healthy vs junk food this week")
plt.ylabel("number of meals")
plt.legend()
plt.tight_layout()
plt.show()
```
