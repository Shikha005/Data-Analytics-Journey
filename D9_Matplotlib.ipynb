```python
import matplotlib.pyplot as plt

subjects = ["maths", "english", "science", "hindi", "computer"]
marks = [78, 85, 92, 70, 88]

plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, color="steelblue")

plt.title("my subject wise marks")
plt.xlabel("subjects")
plt.ylabel("marks")

plt.axhline(y=75, color="red", linestyle="--", label="pass mark")
plt.legend()

plt.tight_layout()
plt.show()
```
