```python
students = {
    "Priya":  85,
    "Rahul":  72,
    "Sneha":  90,
    "Amit":   65,
    "Pooja":  78,
    "Vikram": 55,
    "Neha":   88
}
average = sum(students.values()) / len(students)

topper = max(students, key=students.get)

lowest = min(students, key=students.get)

def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "F - Fail"

print("=" * 35)
print("   STUDENT GRADE REPORT")
print("=" * 35)

for name, marks in students.items():
    grade = get_grade(marks)
    print(f"{name:<10} | Marks: {marks} | Grade: {grade}")

print("=" * 35)
print(f"Class Average : {average:.2f}")
print(f"Topper        : {topper} ({students[topper]})")
print(f"Lowest        : {lowest} ({students[lowest]})")
print("=" * 35)
```
