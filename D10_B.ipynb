```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "month": ["jan", "feb", "mar", "apr", "may", "jun",
               "jul", "aug", "sep", "oct", "nov", "dec"],
    "max_temp": [22, 26, 33, 40, 44, 41, 35, 33, 32, 30, 25, 21],
    "min_temp": [10, 13, 18, 25, 29, 28, 25, 24, 23, 19, 14, 9],
    "rainfall": [5, 8, 3, 2, 15, 95, 280, 310, 180, 25, 10, 4]
}

df = pd.DataFrame(data)

df["avg_temp"] = round((df["max_temp"] + df["min_temp"]) / 2, 1)

print("indore weather 2026")
print("-" * 50)
print(df)
print("-" * 50)
print("hottest month:", df[df["max_temp"] == df["max_temp"].max()]["month"].values[0])
print("coldest month:", df[df["min_temp"] == df["min_temp"].min()]["month"].values[0])
print("most rainfall:", df[df["rainfall"] == df["rainfall"].max()]["month"].values[0])

fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(df["month"], df["max_temp"], color="red", marker="o", label="max temp")
axes[0].plot(df["month"], df["min_temp"], color="blue", marker="o", label="min temp")
axes[0].fill_between(df["month"], df["min_temp"], df["max_temp"], alpha=0.1, color="orange")
axes[0].set_title("indore temperature 2026")
axes[0].set_ylabel("temperature (C)")
axes[0].legend()

axes[1].bar(df["month"], df["rainfall"], color="steelblue")
axes[1].set_title("monthly rainfall")
axes[1].set_ylabel("rainfall (mm)")

plt.tight_layout()
plt.show()
```
