```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print("name:", self.name, "| age:", self.age)

class Doctor(Person):
    def __init__(self, name, age, degree):
        super().__init__(name, age)
        self.degree = degree

    def treat(self):
        print("dr", self.name, "is treating patients")

    def show_degree(self):
        print("degree:", self.degree)

class Teacher:
    def __init__(self, subject, experience):
        self.subject = subject
        self.experience = experience

    def teach(self):
        print(self.name, "is teaching", self.subject)

    def show_exp(self):
        print("experience:", self.experience, "years")

class MedicalTeacher(Doctor, Teacher):
    def __init__(self, name, age, degree, subject, experience):
        Doctor.__init__(self, name, age, degree)
        Teacher.__init__(self, subject, experience)

    def full_profile(self):
        print("=== medical teacher ===")
        self.introduce()
        self.show_degree()
        self.show_exp()

    def daily_work(self):
        self.treat()
        self.teach()

mt = MedicalTeacher("dr sharma", 45, "mbbs md", "anatomy", 15)
mt.full_profile()
print("-" * 25)
mt.daily_work()
```
