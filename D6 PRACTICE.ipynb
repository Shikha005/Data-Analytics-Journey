```python
names = ["dadi", "mummy", "papa", "bhaiya", "me"]
birth_years = [1950, 1975, 1972, 2000, 2005]

current_year = 2026

print("age calculator")
print("-" * 25)

for i in range(len(names)):
    age = current_year - birth_years[i]
    print(names[i], "->", age, "years old")

print("-" * 25)

# found the oldest
oldest_age = 0
oldest_person = ""

for i in range(len(names)):
    age = current_year - birth_years[i]
    if age > oldest_age:
        oldest_age = age
        oldest_person = names[i]

print("oldest:", oldest_person, "->", oldest_age, "years")
```
