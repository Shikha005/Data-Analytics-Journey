```python
numbers = [4, 7, 12, 19, 23, 30, 45, 56, 61, 88]

odd = []
even = []

for n in numbers:
    if n % 2 == 0:
        even.append(n)
    else:
        odd.append(n)

print("numbers i checked:", numbers)
print("-" * 30)
print("even numbers ->", even)
print("odd numbers  ->", odd)
print("-" * 30)
print("total even:", len(even))
print("total odd:", len(odd))
```
