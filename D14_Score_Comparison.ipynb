```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["maths", "english", "science", "hindi", "computer"]
exam1 = [65, 78, 80, 72, 85]
exam2 = [72, 82, 88, 75, 90]

x = np.arange(len(subjects))
width = 0.35

plt.figure(figsize=(9, 5))
plt.bar(x - width/2, exam1, width, label="exam 1", color="coral")
plt.bar(x + width/2, exam2, width, label="exam 2", color="steelblue")

plt.xticks(x, subjects)
plt.title("exam 1 vs exam 2")
plt.ylabel("marks")
plt.legend()
plt.tight_layout()
plt.show()
```
