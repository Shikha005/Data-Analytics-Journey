```python
class Player:
    def bat(self):
        print(self.name, "is batting")

    def bowl(self):
        print(self.name, "is bowling")

class Coach:
    def train(self):
        print(self.name, "is training the team")

    def strategy(self):
        print(self.name, "is planning strategy")

class Manager(Player, Coach):
    def __init__(self, name, team):
        self.name = name
        self.team = team

    def match_day(self):
        print("=== match day ===")
        self.strategy()
        self.bat()
        self.bowl()
        self.train()

m = Manager("rahul dravid", "india")
print("team:", m.team)
m.match_day()
```
