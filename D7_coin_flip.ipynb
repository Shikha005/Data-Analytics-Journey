```python
import random

sides = ["heads", "tails"]

print("coin flipper")
print("-" * 20)

heads = 0
tails = 0

for i in range(10):
    result = random.choice(sides)
    print("flip", i+1, "->", result)

    if result == "heads":
        heads = heads + 1
    else:
        tails = tails + 1

print("-" * 20)
print("heads:", heads)
print("tails:", tails)

if heads > tails:
    print("heads won today!")
elif tails > heads:
    print("tails won today!")
else:
    print("its a tie!")
```
