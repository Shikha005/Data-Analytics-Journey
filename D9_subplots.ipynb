```python
import matplotlib.pyplot as plt

months = ["jan", "feb", "mar", "apr", "may"]
sales = [1200, 1800, 1500, 2200, 1900]
customers = [45, 62, 55, 78, 65]
returns = [5, 8, 4, 10, 7]

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].bar(months, sales, color="steelblue")
axes[0].set_title("monthly sales")
axes[0].set_xlabel("month")
axes[0].set_ylabel("sales (rs)")

axes[1].plot(months, customers, color="green", marker="o")
axes[1].set_title("new customers")
axes[1].set_xlabel("month")
axes[1].set_ylabel("customers")

axes[2].bar(months, returns, color="red")
axes[2].set_title("product returns")
axes[2].set_xlabel("month")
axes[2].set_ylabel("returns")

plt.suptitle("shop performance 2026", fontsize=14)
plt.tight_layout()
plt.show()
```
