```python
import random

students = ["priya", "rahul", "sneha", "amit", "pooja", "vikram", "neha"]

print("random student picker")
print("-" * 25)

picked = random.sample(students, 3)

print("all students:", students)
print("-" * 25)
print("selected students:")
for name in picked:
    print("->", name)

print("-" * 25)

first = random.choice(picked)
print("answers first:", first)
```
