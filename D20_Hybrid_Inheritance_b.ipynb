```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print(self.brand, "is moving")

    def stop(self):
        print(self.brand, "stopped")

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)
        self.doors = doors

    def honk(self):
        print(self.brand, "beep beep!")

class Electric:
    def __init__(self, battery):
        self.battery = battery

    def charge(self):
        self.battery = 100
        print("fully charged!")

    def show_battery(self):
        print("battery:", self.battery, "%")

class ElectricCar(Car, Electric):
    def __init__(self, brand, doors, battery, range_km):
        Car.__init__(self, brand, doors)
        Electric.__init__(self, battery)
        self.range_km = range_km

    def specs(self):
        print("brand:", self.brand)
        print("doors:", self.doors)
        print("battery:", self.battery, "%")
        print("range:", self.range_km, "km")

ec = ElectricCar("tesla", 4, 85, 500)
ec.specs()
ec.move()
ec.honk()
ec.show_battery()
ec.charge()
ec.stop()
```
