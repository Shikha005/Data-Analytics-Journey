```python
import numpy as np

print("different ways to make arrays")
print("-" * 35)

arr1 = np.arange(1, 11)
print("1 to 10:", arr1)

arr2 = np.arange(0, 21, 2)
print("even numbers:", arr2)

arr3 = np.linspace(0, 1, 5)
print("linspace 0 to 1:", arr3)

print("-" * 35)

zeros = np.zeros(5)
ones = np.ones(5)
print("zeros:", zeros)
print("ones:", ones)

print("-" * 35)

import numpy as np
np.random.seed(42)

random_marks = np.random.randint(40, 100, size=10)
print("random marks:", random_marks)
print("average:", round(np.mean(random_marks), 2))
print("passed (>60):", random_marks[random_marks >= 60])
print("failed (<60):", random_marks[random_marks < 60])
```
