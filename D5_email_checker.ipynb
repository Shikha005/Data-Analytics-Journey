```python
emails = [
    "shikha@gmail.com",
    "rahul@yahoo",
    "sneha123@",
    "amit@hotmail.com",
    "pooja.com",
    "test@outlook.com"
]

def check_email(email):
    if "@" not in email:
        return "invalid - no @ found"

    parts = email.split("@")

    if len(parts[0]) == 0:
        return "invalid - nothing before @"

    if "." not in parts[1]:
        return "invalid - no dot after @"

    return "looks valid"

print("email checker")
print("-" * 35)

for email in emails:
    result = check_email(email)
    print(email, "->", result)
```
