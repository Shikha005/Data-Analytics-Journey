```python
players = ["rohit", "virat", "surya", "hardik", "jadeja"]
runs =    [45,       82,      31,       67,       23]
balls =   [38,       61,      28,       42,       19]

def strike_rate(r, b):
    return round((r/b) * 100, 2)

print("cricket score card")
print("-" * 40)

best_score = 0
best_player = ""

for i in range(len(players)):
    sr = strike_rate(runs[i], balls[i])
    print(players[i], "->", runs[i], "runs,", balls[i], "balls, SR:", sr)

    if runs[i] > best_score:
        best_score = runs[i]
        best_player = players[i]

print("-" * 40)
print("top scorer:", best_player, "with", best_score, "runs")

total_runs = 0
for r in runs:
    total_runs = total_runs + r
print("team total:", total_runs, "runs")
```
