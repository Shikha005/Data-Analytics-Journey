```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def study(self):
        print(self.name, "is studying")

class Sports:
    def __init__(self, sport):
        self.sport = sport

    def play(self):
        print(self.name, "plays", self.sport)

class Athlete(Student, Sports):
    def __init__(self, name, marks, sport):
        Student.__init__(self, name, marks)
        Sports.__init__(self, sport)

    def profile(self):
        print("name:", self.name)
        print("marks:", self.marks)
        print("sport:", self.sport)

a = Athlete("priya", 85, "badminton")
a.profile()
a.study()
a.play()
```
