```python
import matplotlib.pyplot as plt

subjects = ["maths", "english", "science", "hindi"]
my_marks =     [78, 85, 90, 72]
friend_marks =  [82, 79, 88, 80]

x = range(len(subjects))

plt.figure(figsize=(8, 5))
plt.plot(x, my_marks, marker="o", label="me", color="blue")
plt.plot(x, friend_marks, marker="o", label="priya", color="red")

plt.xticks(x, subjects)
plt.title("marks comparison - me vs priya")
plt.ylabel("marks")
plt.legend()
plt.tight_layout()
plt.show()
```
