```python
import matplotlib.pyplot as plt

days = ["day1", "day2", "day3", "day4", "day5", "day6", "day7"]
km = [2.0, 2.5, 2.0, 3.0, 2.5, 3.5, 4.0]

plt.figure(figsize=(8, 5))
plt.plot(days, km, marker="o", color="green", linewidth=2)
plt.fill_between(days, km, alpha=0.1, color="green")

plt.title("my running progress this week")
plt.ylabel("kilometers")
plt.axhline(y=3, color="red", linestyle="--", label="daily goal")
plt.legend()
plt.tight_layout()
plt.show()
```
