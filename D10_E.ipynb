```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "name": ["priya", "rahul", "sneha", "amit", "pooja", "vikram", "neha"],
    "maths": [78, 65, 90, 55, 82, 48, 88],
    "english": [85, 72, 88, 60, 79, 55, 92],
    "science": [92, 68, 95, 58, 85, 50, 89]
}

df = pd.DataFrame(data)

df["total"] = df["maths"] + df["english"] + df["science"]
df["percentage"] = round((df["total"] / 300) * 100, 2)

print("student report")
print("-" * 50)
print(df)
print("-" * 50)
print("class topper:", df[df["total"] == df["total"].max()]["name"].values[0])
print("class average:", round(df["percentage"].mean(), 2), "%")

plt.figure(figsize=(9, 5))
plt.bar(df["name"], df["percentage"], color="steelblue")
plt.axhline(y=60, color="red", linestyle="--", label="pass line")
plt.title("student percentage comparison")
plt.xlabel("students")
plt.ylabel("percentage %")
plt.legend()
plt.tight_layout()
plt.show()
```
