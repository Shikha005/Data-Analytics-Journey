```python
class Phone:
    def __init__(self, brand):
        self.brand = brand

    def call(self):
        print(self.brand, "is calling")

class Camera:
    def __init__(self, mp):
        self.mp = mp

    def click(self):
        print("photo clicked in", self.mp, "mp")

class Smartphone(Phone, Camera):
    def __init__(self, brand, mp, storage):
        Phone.__init__(self, brand)
        Camera.__init__(self, mp)
        self.storage = storage

    def specs(self):
        print("brand:", self.brand)
        print("camera:", self.mp, "mp")
        print("storage:", self.storage, "gb")

s = Smartphone("samsung", 108, 256)
s.specs()
s.call()
s.click()
```
