```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "team": ["mi", "rcb", "csk", "kkr", "dc", "srh", "rr", "pbks"],
    "matches": [14, 14, 14, 14, 14, 14, 14, 14],
    "wins": [9, 7, 8, 6, 5, 7, 8, 4],
    "runs_scored": [2180, 1950, 2050, 1880, 1790, 1920, 2010, 1760],
    "wickets": [165, 148, 158, 142, 135, 152, 160, 130]
}

df = pd.DataFrame(data)

df["losses"] = df["matches"] - df["wins"]
df["win_percent"] = round((df["wins"] / df["matches"]) * 100, 1)
df["avg_runs"] = round(df["runs_scored"] / df["matches"], 1)

print("ipl 2026 standings")
print("-" * 60)
print(df[["team", "wins", "losses", "win_percent", "avg_runs"]])
print("-" * 60)
print("best team:", df[df["wins"] == df["wins"].max()]["team"].values[0])
print("highest avg runs:", df[df["avg_runs"] == df["avg_runs"].max()]["team"].values[0])

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].bar(df["team"], df["wins"], color="steelblue")
axes[0].set_title("wins per team")
axes[0].set_ylabel("wins")

axes[1].plot(df["team"], df["win_percent"], color="green", marker="o")
axes[1].set_title("win percentage")
axes[1].set_ylabel("win %")
axes[1].axhline(y=50, color="red", linestyle="--", label="50% line")
axes[1].legend()

plt.suptitle("ipl 2026 team analysis")
plt.tight_layout()
plt.show()
```
