```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def study(self):
        print(self.name, "is studying")

    def show_marks(self):
        print("marks:", self.marks)

class Sports:

    def __init__(self, sport, medals):
        self.sport = sport
        self.medals = medals

    def play(self):
        print(self.name, "is playing", self.sport)

    def show_medals(self):
        print("medals won:", self.medals)

class Athlete(Student, Sports):

    def __init__(self, name, marks, sport, medals):
        Student.__init__(self, name, marks)
        Sports.__init__(self, sport, medals)

    def profile(self):
        print("=== athlete profile ===")
        print("name:", self.name)
        print("marks:", self.marks)
        print("sport:", self.sport)
        print("medals:", self.medals)

# testing
a1 = Athlete("priya", 85, "badminton", 3)
a2 = Athlete("rahul", 72, "cricket", 1)

a1.profile()
a1.study()
a1.play()
a1.show_marks()
a1.show_medals()

print("-" * 25)
a2.profile()
```
