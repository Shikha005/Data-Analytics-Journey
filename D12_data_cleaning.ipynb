```python
import pandas as pd
import numpy as np

data = {
    "name": ["Priya", "rahul", "  Sneha  ", "amit", None, "POOJA", "vikram", None],
    "age": [20, 22, 19, None, 20, 23, 19, 22],
    "city": ["Indore", "bhopal", "DELHI", "mumbai", "indore", None, "delhi", "bhopal"],
    "marks": [85, 72, 90, 65, None, 78, 55, 88],
    "phone": ["9876543210", "98123", "9856781234", "9890123456",
               "9812341234", "9856785678", "abc", "9834567890"]
}

df = pd.DataFrame(data)

print("original messy data:")
print(df)
print("-" * 40)
print("null values:")
print(df.isnull().sum())
print("-" * 40)

df = df.dropna(subset=["name"])
print("after dropping missing names:", df.shape)

avg_age = round(df["age"].mean())
df["age"] = df["age"].fillna(avg_age)

median_marks = df["marks"].median()
df["marks"] = df["marks"].fillna(median_marks)

df["city"] = df["city"].fillna("unknown")

df["name"] = df["name"].str.strip()
df["name"] = df["name"].str.lower()

df["city"] = df["city"].str.lower()

df["phone_valid"] = df["phone"].apply(lambda x: "valid" if len(str(x)) == 10 and str(x).isdigit() else "invalid")

print("cleaned data:")
print(df)
print("-" * 40)
print("null values after cleaning:")
print(df.isnull().sum())
```
