```python
import matplotlib.pyplot as plt

cities = ["indore", "bhopal", "delhi", "mumbai", "pune"]
students = [45, 30, 80, 65, 25]

plt.figure(figsize=(8, 5))
plt.bar(cities, students, color=["steelblue", "coral", "green", "purple", "orange"])

plt.title("students per city")
plt.xlabel("city")
plt.ylabel("number of students")

for i in range(len(cities)):
    plt.text(i, students[i] + 1, str(students[i]), ha="center")

plt.tight_layout()
plt.show()
```
