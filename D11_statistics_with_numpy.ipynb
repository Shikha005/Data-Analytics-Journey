```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
scores = np.random.randint(35, 100, size=50)

print("class exam analysis")
print("-" * 35)
print("scores:", scores)
print("-" * 35)

print("total students:", len(scores))
print("average score:", round(np.mean(scores), 2))
print("median score:", np.median(scores))
print("highest:", np.max(scores))
print("lowest:", np.min(scores))
print("std deviation:", round(np.std(scores), 2))

print("-" * 35)

passed = scores[scores >= 60]
failed = scores[scores < 60]
print("passed:", len(passed))
print("failed:", len(failed))
print("pass percentage:", round(len(passed)/len(scores)*100, 1), "%")

print("-" * 35)

print("25th percentile:", np.percentile(scores, 25))
print("75th percentile:", np.percentile(scores, 75))

plt.figure(figsize=(8, 5))
plt.hist(scores, bins=10, color="steelblue", edgecolor="black")
plt.axvline(x=60, color="red", linestyle="--", label="pass mark")
plt.axvline(x=np.mean(scores), color="green", linestyle="--", label="average")
plt.title("score distribution of 50 students")
plt.xlabel("scores")
plt.ylabel("number of students")
plt.legend()
plt.tight_layout()
plt.show()
```
