```python
import matplotlib.pyplot as plt

subjects = ["maths", "english", "science", "hindi", "computer"]
rating = [5, 8, 9, 7, 10]

plt.figure(figsize=(8, 5))
plt.barh(subjects, rating, color="steelblue")

plt.title("how much i like each subject")
plt.xlabel("rating out of 10")
plt.axvline(x=7, color="red", linestyle="--", label="average line")
plt.legend()
plt.tight_layout()
plt.show()
```
