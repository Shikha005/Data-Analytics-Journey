```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        result = round(3.14 * self.radius ** 2, 2)
        print("circle area:", result)

class Rectangle:
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        result = self.l * self.w
        print("rectangle area:", result)

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        result = 0.5 * self.base * self.height
        print("triangle area:", result)

shapes = [Circle(7), Rectangle(5, 3), Triangle(6, 4)]

for shape in shapes:
    shape.area()
```
