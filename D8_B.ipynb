```python
import pandas as pd

data = {
    "name": ["priya", "rahul", "sneha", "amit", "pooja", "vikram", "neha"],
    "city": ["indore", "bhopal", "delhi", "mumbai", "indore", "delhi", "bhopal"],
    "marks": [85, 72, 90, 65, 78, 55, 88],
    "age": [20, 22, 19, 21, 20, 23, 19]
}

df = pd.DataFrame(data)

print("all students:")
print(df)
print("-" * 35)

passed = df[df["marks"] > 75]
print("students with marks > 75:")
print(passed)
print("-" * 35)

indore = df[df["city"] == "indore"]
print("students from indore:")
print(indore)
print("-" * 35)

good_indore = df[(df["marks"] > 70) & (df["city"] == "indore")]
print("indore students with marks > 70:")
print(good_indore)
```
