```python
class Father:
    def coding(self):
        print("got coding skills from father")

    def show_height(self):
        print("height:", self.height, "cm")

class Mother:
    def cooking(self):
        print("got cooking skills from mother")

    def show_hobby(self):
        print("hobby:", self.hobby)

class Child(Father, Mother):
    def __init__(self, name, height, hobby):
        self.name = name
        self.height = height
        self.hobby = hobby

    def introduce(self):
        print("name:", self.name)
        self.show_height()
        self.show_hobby()

c = Child("shikha", 162, "painting")
c.introduce()
c.coding()
c.cooking()
```
