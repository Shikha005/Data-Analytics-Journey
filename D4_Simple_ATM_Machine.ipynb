```python
balance = 10000

def check_balance():
    print("current balance: Rs.", balance)

def deposit(amount):
    global balance
    if amount > 0:
        balance = balance + amount
        print("Rs.", amount, "added")
        print("new balance: Rs.", balance)
    else:
        print("invalid amount")

def withdraw(amount):
    global balance
    if amount > balance:
        print("not enough balance")
    elif amount <= 0:
        print("invalid amount")
    else:
        balance = balance - amount
        print("Rs.", amount, "withdrawn")
        print("remaining balance: Rs.", balance)

print("welcome to my atm")
print("-" * 25)
check_balance()
print("-" * 25)
deposit(2000)
print("-" * 25)
withdraw(500)
print("-" * 25)
withdraw(15000)
print("-" * 25)
check_balance()
```
