```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "study_hours": [1, 2, 3, 4, 5, 6, 7, 8, 3, 5, 4, 6],
    "marks": [40, 55, 60, 70, 75, 85, 88, 95, 58, 72, 68, 80],
    "gender": ["f", "m", "f", "m", "f", "f", "m", "f", "m", "f", "m", "f"]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.scatterplot(x="study_hours", y="marks", hue="gender", data=df, s=100)

plt.title("study hours vs marks")
plt.xlabel("hours studied")
plt.ylabel("marks scored")
plt.tight_layout()
plt.show()
```
