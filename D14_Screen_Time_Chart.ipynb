```python
import matplotlib.pyplot as plt

apps = ["instagram", "youtube", "whatsapp", "chrome", "netflix"]
hours = [3.5, 2.0, 1.5, 1.0, 2.5]
colors = ["#E1306C", "#FF0000", "#25D366", "#4285F4", "#E50914"]

plt.figure(figsize=(7, 7))
plt.pie(hours, labels=apps, colors=colors,
        autopct="%1.0f%%", startangle=90)

plt.title("my daily screen time")
plt.tight_layout()
plt.show()
```
