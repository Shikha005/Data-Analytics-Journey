```python
import matplotlib.pyplot as plt

months = ["jan", "feb", "mar", "apr", "may", "jun"]
my_savings = [500, 800, 600, 1200, 900, 1500]
my_spending = [1200, 900, 1400, 800, 1100, 700]

plt.figure(figsize=(9, 5))

plt.plot(months, my_savings, color="green", marker="o", label="savings")
plt.plot(months, my_spending, color="red", marker="o", label="spending")

plt.title("my savings vs spending - 2026")
plt.xlabel("month")
plt.ylabel("amount (rs)")
plt.legend()

plt.tight_layout()
plt.show()
```
