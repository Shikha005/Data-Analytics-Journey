```python
from abc import ABC, abstractmethod

class Payment(ABC):

    def __init__(self, amount):
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass

    @abstractmethod
    def confirm(self):
        pass

    def receipt(self):
        print("amount paid: Rs.", self.amount)

class UPI(Payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)
        self.upi_id = upi_id

    def pay(self):
        print("paying Rs.", self.amount, "via upi:", self.upi_id)

    def confirm(self):
        print("upi payment confirmed!")

class Card(Payment):
    def __init__(self, amount, last4):
        super().__init__(amount)
        self.last4 = last4

    def pay(self):
        print("paying Rs.", self.amount, "via card ending", self.last4)

    def confirm(self):
        print("card payment confirmed!")

u = UPI(500, "shikha@upi")
c = Card(1200, "4321")

u.pay()
u.confirm()
u.receipt()

print("-" * 25)

c.pay()
c.confirm()
c.receipt()
```
