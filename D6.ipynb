```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    try:
        return round(a / b, 2)
    except ZeroDivisionError:
        return "error: cant divide by zero"

calculations = [
    (10, 5),
    (100, 0),
    (7, 3),
    (50, 4),
    (0, 8)
]

print("calculator results")
print("-" * 35)

for a, b in calculations:
    print(f"{a} + {b} =", add(a, b))
    print(f"{a} - {b} =", subtract(a, b))
    print(f"{a} x {b} =", multiply(a, b))
    print(f"{a} / {b} =", divide(a, b))
    print("-" * 35)
```
