```python
class Food:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def prepare(self):
        print("preparing", self.name)

    def info(self):
        print("-" * 25)
        print("item:", self.name)
        print("price: Rs.", self.price)
        self.prepare()

class Pizza(Food):
    def prepare(self):
        print("baking pizza in oven for 20 mins")

class Biryani(Food):
    def prepare(self):
        print("slow cooking biryani for 40 mins")

class Burger(Food):
    def prepare(self):
        print("grilling burger patty for 10 mins")

class IceCream(Food):
    def prepare(self):
        print("scooping ice cream, no cooking needed!")

order = [
    Pizza("margherita", 250),
    Biryani("chicken biryani", 320),
    Burger("veg burger", 150),
    IceCream("chocolate scoop", 80)
]

for item in order:
    item.info()

print("-" * 25)
print("order complete!")
```
