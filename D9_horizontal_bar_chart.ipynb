```python
import matplotlib.pyplot as plt

players = ["rohit sharma", "virat kohli", "suryakumar", "shubman gill", "kl rahul"]
runs = [520, 680, 410, 590, 350]

plt.figure(figsize=(9, 5))

bars = plt.barh(players, runs, color="coral")

for bar, run in zip(bars, runs):
    plt.text(
        bar.get_width() + 5,
        bar.get_y() + bar.get_height()/2,
        str(run),
        va="center"
    )

plt.title("ipl top scorers 2026")
plt.xlabel("runs scored")
plt.tight_layout()
plt.show()
```
