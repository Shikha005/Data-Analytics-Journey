```python
import numpy as np
import matplotlib.pyplot as plt

products = ["chips", "biscuits", "chocolate", "juice", "water", "colddrink"]
prices = np.array([20, 30, 50, 60, 20, 40])
stock = np.array([150, 200, 80, 90, 300, 180])
sold = np.array([120, 180, 65, 72, 250, 160])

remaining = stock - sold
revenue = prices * sold
profit = revenue * 0.20
print("shop inventory report")
print("-" * 55)

for i in range(len(products)):
    print(products[i])
    print("  sold:", sold[i], "| remaining:", remaining[i], "| revenue: Rs.", revenue[i])

print("-" * 55)
print("total revenue: Rs.", np.sum(revenue))
print("total profit: Rs.", round(np.sum(profit), 2))
print("best seller:", products[np.argmax(sold)])
print("worst seller:", products[np.argmin(sold)])

print("-" * 55)
print("low stock items (less than 30 remaining):")
low_stock = np.array(products)[remaining < 30]
if len(low_stock) > 0:
    for item in low_stock:
        print("->", item, "- reorder needed!")
else:
    print("all items have good stock")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(products, revenue, color="steelblue")
axes[0].set_title("revenue by product")
axes[0].set_ylabel("revenue (rs)")
axes[0].tick_params(axis="x", rotation=45)

axes[1].bar(products, remaining, color="coral")
axes[1].axhline(y=30, color="red", linestyle="--", label="reorder level")
axes[1].set_title("remaining stock")
axes[1].set_ylabel("units")
axes[1].legend()
axes[1].tick_params(axis="x", rotation=45)

plt.suptitle("shop inventory dashboard")
plt.tight_layout()
plt.show()
```
