```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "category": ["food", "transport", "recharge", "stationery", "entertainment", "savings"],
    "planned": [800, 300, 199, 150, 200, 500],
    "actual": [950, 250, 199, 180, 350, 300]
}

df = pd.DataFrame(data)

df["difference"] = df["actual"] - df["planned"]
df["status"] = df["difference"].apply(lambda x: "overspent" if x > 0 else "saved")

print("monthly budget report")
print("-" * 55)
print(df)
print("-" * 55)
print("total planned:", df["planned"].sum())
print("total actual:", df["actual"].sum())
print("total overspent:", df[df["difference"] > 0]["difference"].sum())

fig, ax = plt.subplots(figsize=(10, 5))

x = range(len(df["category"]))
width = 0.35

ax.bar([i - width/2 for i in x], df["planned"], width, label="planned", color="green")
ax.bar([i + width/2 for i in x], df["actual"], width, label="actual", color="coral")

ax.set_xticks(x)
ax.set_xticklabels(df["category"])
ax.set_title("planned vs actual spending")
ax.set_ylabel("amount (rs)")
ax.legend()

plt.tight_layout()
plt.show()
```
