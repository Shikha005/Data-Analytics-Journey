```python
import matplotlib.pyplot as plt

days = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
hours = [2, 3, 1.5, 2.5, 3, 4, 1]

plt.figure(figsize=(8, 5))
bars = plt.bar(days, hours, color="steelblue")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.05,
        str(bar.get_height()),
        ha="center"
    )

plt.title("my study hours this week")
plt.ylabel("hours")
plt.axhline(y=2, color="red", linestyle="--", label="daily goal")
plt.legend()
plt.tight_layout()
plt.show()
```
