```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

cities = ["Delhi", "Mumbai", "Shimla", "Chennai", "Manali"]
temps_celsius = [42, 35, 18, 38, 8]

print("city temperatures today")
print("-" * 35)

for i in range(len(cities)):
    c = temps_celsius[i]
    f = celsius_to_fahrenheit(c)
    print(cities[i], "->", c, "C =", round(f, 1), "F")

print("-" * 35)

hottest = max(temps_celsius)
coldest = min(temps_celsius)
print("hottest city temp:", hottest, "C")
print("coldest city temp:", coldest, "C")
```
