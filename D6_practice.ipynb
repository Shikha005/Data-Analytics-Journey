```python
foods = ["maggi", "pizza", "biryani", "momos", "pasta"]
prices = [20, 200, 150, 30, 120]

print("my favourite foods")
print("-" * 25)

for i in range(len(foods)):
    print(foods[i], "-> Rs.", prices[i])

print("-" * 25)

cheapest = min(prices)
expensive = max(prices)

print("cheapest:", cheapest)
print("most expensive:", expensive)
```
