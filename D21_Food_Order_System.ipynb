```python
from abc import ABC, abstractmethod

class FoodItem(ABC):

    def __init__(self, name, price):
        self.name = name
        self.price = price

    @abstractmethod
    def prepare(self):
        pass

    @abstractmethod
    def time_needed(self):
        pass

    def order_summary(self):
        print("item:", self.name)
        print("price: Rs.", self.price)
        print("time:", self.time_needed(), "mins")

class Pizza(FoodItem):
    def prepare(self):
        print("making pizza dough, adding toppings, baking")

    def time_needed(self):
        return 20

class Biryani(FoodItem):
    def prepare(self):
        print("soaking rice, cooking meat, dum cooking")

    def time_needed(self):
        return 35

class Burger(FoodItem):
    def prepare(self):
        print("grilling patty, toasting bun, assembling")

    def time_needed(self):
        return 10

orders = [
    Pizza("margherita", 250),
    Biryani("chicken biryani", 320),
    Burger("aloo burger", 150)
]

for item in orders:
    item.order_summary()
    item.prepare()
    print("-" * 30)
```
