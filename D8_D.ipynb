```python
import pandas as pd

data = {
    "player": ["rohit", "virat", "surya", "hardik", "gill", "pant"],
    "runs": [450, 620, 380, 290, 540, 410],
    "matches": [12, 15, 11, 10, 14, 13],
    "centuries": [2, 3, 1, 0, 2, 1]
}

df = pd.DataFrame(data)

print("ipl stats")
print("-" * 40)
print(df)
print("-" * 40)

print("top scorers:")
sorted_df = df.sort_values("runs", ascending=False)
print(sorted_df)
print("-" * 40)

df["avg"] = round(df["runs"] / df["matches"], 2)
print("with average column added:")
print(df)
print("-" * 40)

print("top 3 players:")
print(df.sort_values("runs", ascending=False).head(3))
```
