```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "maths":    [78, 65, 90, 55, 82, 48, 88, 72],
    "english":  [85, 72, 88, 60, 79, 55, 92, 68],
    "science":  [92, 68, 95, 58, 85, 50, 89, 74],
    "hindi":    [70, 75, 85, 62, 77, 60, 80, 65],
    "computer": [88, 60, 92, 50, 85, 45, 95, 70]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

plt.title("subject marks correlation")
plt.tight_layout()
plt.show()
```
