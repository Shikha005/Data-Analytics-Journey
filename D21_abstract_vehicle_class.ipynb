```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def __init__(self, brand, fuel):
        self.brand = brand
        self.fuel = fuel

    @abstractmethod
    def speed(self):
        pass

    @abstractmethod
    def sound(self):
        pass

    def info(self):
        print("brand:", self.brand)
        print("fuel:", self.fuel)

class Bike(Vehicle):
    def speed(self):
        print(self.brand, "max speed: 120 kmph")

    def sound(self):
        print(self.brand, "makes: vroom vroom")

class Truck(Vehicle):
    def speed(self):
        print(self.brand, "max speed: 80 kmph")

    def sound(self):
        print(self.brand, "makes: dhad dhad")

class Cycle(Vehicle):
    def speed(self):
        print(self.brand, "max speed: 25 kmph")

    def sound(self):
        print(self.brand, "makes: no sound lol")

b = Bike("bajaj", "petrol")
t = Truck("tata", "diesel")
cy = Cycle("hero", "none")

for v in [b, t, cy]:
    v.info()
    v.speed()
    v.sound()
    print("-" * 25)
```
