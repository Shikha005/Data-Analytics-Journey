```python
import numpy as np

normal_list = [1, 2, 3, 4, 5]

np_array = np.array([1, 2, 3, 4, 5])

print("normal list:", normal_list)
print("numpy array:", np_array)
print("-" * 30)

print("list * 2:", [x * 2 for x in normal_list])
print("array * 2:", np_array * 2)

print("-" * 30)

marks = np.array([78, 85, 92, 65, 88, 72, 90])

print("marks:", marks)
print("average:", np.mean(marks))
print("highest:", np.max(marks))
print("lowest:", np.min(marks))
print("total:", np.sum(marks))
print("std deviation:", round(np.std(marks), 2))

print("-" * 30)
print("marks above 80:", marks[marks > 80])
```
