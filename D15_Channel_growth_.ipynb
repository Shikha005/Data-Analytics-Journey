```python
import matplotlib.pyplot as plt

months = ["jan", "feb", "mar", "apr", "may", "jun"]
subscribers = [100, 250, 480, 820, 1200, 1800]

plt.figure(figsize=(8, 5))
plt.plot(months, subscribers, marker="o", color="red", linewidth=2)
plt.fill_between(months, subscribers, alpha=0.1, color="red")

for i in range(len(months)):
    plt.text(i, subscribers[i] + 30, str(subscribers[i]), ha="center")

plt.title("my youtube channel growth")
plt.ylabel("subscribers")
plt.tight_layout()
plt.show()
```
