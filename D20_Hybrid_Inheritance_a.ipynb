```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print("hi i am", self.name, "age", self.age)

class Employee(Person):
    def __init__(self, name, age, company):
        super().__init__(name, age)
        self.company = company

    def work(self):
        print(self.name, "is working at", self.company)

class Student:
    def __init__(self, college, year):
        self.college = college
        self.year = year

    def study(self):
        print(self.name, "is studying at", self.college)

class Intern(Employee, Student):
    def __init__(self, name, age, company, college, year, stipend):
        Employee.__init__(self, name, age, company)
        Student.__init__(self, college, year)
        self.stipend = stipend

    def profile(self):
        print("=== intern profile ===")
        print("name:", self.name)
        print("company:", self.company)
        print("college:", self.college)
        print("year:", self.year)
        print("stipend: Rs.", self.stipend)

i = Intern("priya", 21, "tcs", "ips indore", 3, 15000)
i.profile()
i.introduce()
i.work()
i.study()
```
