```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "subject": ["maths"]*6 + ["english"]*6 + ["science"]*6,
    "marks": [55, 65, 72, 78, 85, 90,
              60, 70, 75, 80, 88, 92,
              50, 62, 70, 82, 88, 95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.boxplot(x="subject", y="marks", data=df, palette="Set2")

plt.title("marks distribution by subject")
plt.tight_layout()
plt.show()
```
