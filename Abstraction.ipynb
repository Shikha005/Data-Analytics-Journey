```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def describe(self):
        print("i am a shape")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return round(3.14 * self.radius ** 2, 2)

    def perimeter(self):
        return round(2 * 3.14 * self.radius, 2)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

c = Circle(7)
r = Rectangle(5, 3)

print("circle:")
print("area:", c.area())
print("perimeter:", c.perimeter())
c.describe()

print("-" * 25)
print("rectangle:")
print("area:", r.area())
print("perimeter:", r.perimeter())
r.describe()
```
