```python
import matplotlib.pyplot as plt

months = ["jan", "feb", "mar", "apr", "may", "jun",
          "jul", "aug", "sep", "oct", "nov", "dec"]
temp = [22, 26, 33, 40, 44, 41, 35, 33, 32, 30, 25, 21]

plt.figure(figsize=(10, 5))
plt.plot(months, temp, marker="o", color="red", linewidth=2)

plt.fill_between(months, temp, alpha=0.1, color="red")

plt.title("indore temperature 2026")
plt.xlabel("month")
plt.ylabel("temperature (celsius)")

plt.axhline(y=35, color="orange", linestyle="--", label="heat warning")
plt.legend()
plt.tight_layout()
plt.show()
```
