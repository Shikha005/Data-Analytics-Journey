```python
import pandas as pd

data = {
    "subject": ["maths", "english", "science", "hindi", "computer", "sst"],
    "my_marks": [78, 85, 92, 70, 88, 75],
    "class_avg": [65, 72, 80, 68, 75, 70],
    "highest": [95, 90, 98, 85, 95, 88]
}

df = pd.DataFrame(data)

print("my marks vs class")
print("-" * 35)
print(df)
print("-" * 35)

print("my total:", df["my_marks"].sum())
print("class avg total:", df["class_avg"].sum())
print("-" * 35)

print("my average:", round(df["my_marks"].mean(), 2))
print("class average:", round(df["class_avg"].mean(), 2))
print("-" * 35)

best = df[df["my_marks"] == df["my_marks"].max()]
worst = df[df["my_marks"] == df["my_marks"].min()]

print("best subject:", best["subject"].values[0])
print("worst subject:", worst["subject"].values[0])

print("-" * 35)
print("full stats:")
print(df["my_marks"].describe())
```
