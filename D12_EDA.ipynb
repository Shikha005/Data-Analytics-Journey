```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "name": ["priya", "rahul", "sneha", "amit", "pooja",
             "vikram", "neha", "raj", "simran", "arjun"],
    "city": ["indore", "bhopal", "delhi", "mumbai", "indore",
              "delhi", "mumbai", "bhopal", "indore", "delhi"],
    "age": [20, 22, 19, 21, 20, 23, 19, 22, 20, 21],
    "marks": [78, 65, 90, 55, 82, 48, 88, 72, 95, 60],
    "fees_paid": [True, False, True, True, False, True, True, False, True, True]
}

df = pd.DataFrame(data)

print("=== EXPLORATORY DATA ANALYSIS ===")
print("-" * 40)

print("rows:", df.shape[0])
print("columns:", df.shape[1])

print("\ndata types:")
print(df.dtypes)

print("\nmissing values:")
print(df.isnull().sum())

print("\nunique cities:", df["city"].unique())
print("city counts:")
print(df["city"].value_counts())

print("\nmarks stats:")
print("mean:", df["marks"].mean())
print("median:", df["marks"].median())
print("std:", round(df["marks"].std(), 2))

print("\nfees status:")
print(df["fees_paid"].value_counts())

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].hist(df["marks"], bins=5, color="steelblue", edgecolor="black")
axes[0, 0].set_title("marks distribution")

city_counts = df["city"].value_counts()
axes[0, 1].bar(city_counts.index, city_counts.values, color="coral")
axes[0, 1].set_title("students per city")

axes[1, 0].hist(df["age"], bins=4, color="green", edgecolor="black")
axes[1, 0].set_title("age distribution")

fees = df["fees_paid"].value_counts()
axes[1, 1].pie(fees.values, labels=["paid", "not paid"], autopct="%1.0f%%")
axes[1, 1].set_title("fees status")

plt.suptitle("student data EDA")
plt.tight_layout()
plt.show()
```
