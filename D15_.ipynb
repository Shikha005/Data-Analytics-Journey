```python
import matplotlib.pyplot as plt

days = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
hours = [6, 5, 7, 4, 6, 9, 8]

colors = ["red" if h < 6 else "orange" if h < 8 else "green" for h in hours]

plt.figure(figsize=(8, 5))
plt.bar(days, hours, color=colors)

plt.axhline(y=8, color="blue", linestyle="--", label="ideal sleep")
plt.title("my sleep tracker this week")
plt.ylabel("hours slept")
plt.legend()
plt.tight_layout()
plt.show()
```
