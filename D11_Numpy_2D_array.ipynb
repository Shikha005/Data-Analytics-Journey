```python
import numpy as np

marks = np.array([
    [78, 85, 92],   # priya
    [65, 72, 68],   # rahul
    [90, 88, 95],   # sneha
    [55, 60, 58]    # amit
])

students = ["priya", "rahul", "sneha", "amit"]
subjects = ["maths", "english", "science"]

print("marks table")
print("-" * 30)
print(marks)
print("-" * 30)

print("shape:", marks.shape)   # 4 rows, 3 columns
print("total elements:", marks.size)

print("-" * 30)

print("student totals:")
for i in range(len(students)):
    total = np.sum(marks[i])
    avg = round(np.mean(marks[i]), 1)
    print(students[i], "-> total:", total, "avg:", avg)

print("-" * 30)
print("subject averages:")
for i in range(len(subjects)):
    avg = round(np.mean(marks[:, i]), 1)
    print(subjects[i], "->", avg)
```
