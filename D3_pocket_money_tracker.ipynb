```python
pocket_money = 1500

spent = {
    "food": 250,
    "recharge": 199,
    "stationery": 80,
    "random stuff": 150
}

print("my pocket money tracker")
print("-" * 30)
print("total money:", pocket_money)
print("-" * 30)

total_spent = 0

for item in spent:
    print("spent on", item, "->", spent[item])
    total_spent = total_spent + spent[item]

remaining = pocket_money - total_spent

print("-" * 30)
print("total spent:", total_spent)
print("money left:", remaining)

if remaining < 300:
    print("warning: running low on money!")
elif remaining < 600:
    print("okay but be careful now")
else:
    print("good, still have enough")
```
