```python
sentence = "data analytics is really interesting and i want to learn data science after data analytics"

words = sentence.split()

print("original sentence:")
print(sentence)
print("-" * 40)
print("total words:", len(words))
print("-" * 40)

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

print("word frequencies:")
for word in word_count:
    if word_count[word] > 1:
        print(word, "->", word_count[word], "times")
```
