```python
import matplotlib.pyplot as plt

players = ["rohit", "virat", "gill", "surya", "hardik"]
runs = [520, 680, 590, 410, 320]

plt.figure(figsize=(8, 5))
bars = plt.bar(players, runs, color="steelblue")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 5,
        str(bar.get_height()),
        ha="center"
    )

plt.title("ipl 2026 top scorers")
plt.ylabel("runs")
plt.tight_layout()
plt.show()
```
