```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print(self.brand, "moving at", self.speed, "kmph")

    def stop(self):
        print(self.brand, "stopped")

class Car(Vehicle):
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)
        self.doors = doors

    def honk(self):
        print(self.brand, "beep beep!")

    def info(self):
        print("brand:", self.brand)
        print("speed:", self.speed, "kmph")
        print("doors:", self.doors)

class ElectricCar(Car):
    def __init__(self, brand, speed, doors, battery):
        super().__init__(brand, speed, doors)
        self.battery = battery

    def charge(self):
        self.battery = 100
        print(self.brand, "fully charged!")

    def show_battery(self):
        print("battery:", self.battery, "%")

ec = ElectricCar("tesla", 200, 4, 80)
ec.info()
ec.move()
ec.honk()
ec.show_battery()
ec.charge()
ec.show_battery()
ec.stop()
```
